```python
import pandas as pd

query_df = pd.read_csv("data/ids/query_enriched_ids.csv")
kaggle_df = pd.read_csv("data/ids/us_canada_under60s_videos.csv")

print(f"query:  {len(query_df)}Í∞ú  |  columns: {query_df.columns.tolist()}")
print(f"kaggle: {len(kaggle_df)}Í∞ú  |  columns: {kaggle_df.columns.tolist()}")
```

```python
common_cols = list(set(query_df.columns) & set(kaggle_df.columns))
print("Í≥µÌÜµ Ïª¨Îüº:", common_cols)

combined = pd.concat(
    [query_df[common_cols], kaggle_df[common_cols]], ignore_index=True
).drop_duplicates(subset="video_id")

print(f"\nÌï©ÏÇ∞: {len(combined)}Í∞ú (Ï§ëÎ≥µ Ï†úÍ±∞ ÌõÑ)")
combined.head()
```

```python
combined.to_csv("data/ids/combined_ids.csv", index=False)
print(f"Ï†ÄÏû• ÏôÑÎ£å ‚Üí data/ids/combined_ids.csv")
```
